```python
#Check config dta
f = open('config.txt', 'r')
print('------------------------------ ')
print('---check config data---')
bombconfig = []

while True:
    s = f.readline()

    if s == '': # check file end
        break

    # spliting line to key and value
    d = s.rstrip().split('=');
    print(d[0] + ': ' + d[1])
    bombconfig.append(d[1])
f.close()

print('\n')
print(f'configdata : \n {bombconfig}\n')
file_path = bombconfig[0].strip()
pic_path = bombconfig[1].strip()
times = bombconfig[2]
delay_time = bombconfig[3]
url = bombconfig[4]

print(f'\n data : \n {file_path}\n{pic_path}\n{times}\n{delay_time}\n{url}\n')

print('\nchecked config data')
print('------------------------------ ')



#Check connect internet
print('------------------------------ ')
print('---check connect url---\n')


import requests
print(f'{url}')
res = requests.get(url)
#res.encoding = "utf-8"
print(res)
if res.status_code == 200:
    print("Successful")
elif res.status_code == 404:
    print("Error 404 page not found")
else:
    print("Not both 200 and 404")
    
    
print('------------------------------ ')
```
